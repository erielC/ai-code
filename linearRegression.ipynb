```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
rng = np.random.default_rng()

x = rng.random(10)
y = rng.random(10)

print(x)
print(y)

linReg = stats.linregress(x, y)
print(linReg)

plt.plot(x, y, 'o', label='original data')
plt.plot(x, linReg.intercept + linReg.slope*x, 'r', label='fitted line')
plt.legend()
plt.show()
```
